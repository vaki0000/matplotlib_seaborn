```python
titanic = sns.load_dataset("titanic")
titanic
```

```python
sns.catplot(x="pclass", y="survived", data=titanic, kind="bar");
```

```python
g = sns.PairGrid(tips)
g.map(sns.scatterplot);
```

```python
g = sns.PairGrid(tips)
g.map_upper(sns.scatterplot)
g.map_diag(sns.histplot);
```

```python
g = sns.PairGrid(tips)
g.map_upper(sns.scatterplot)
g.map_diag(sns.histplot)
g.map_lower(sns.kdeplot, colors="red");
```

```python
g = sns.PairGrid(tips, hue="sex", palette="magma")
g.map_upper(sns.scatterplot, s=100, edgecolor="black")
g.map_diag(sns.histplot)
g.map_lower(sns.kdeplot)
g.add_legend();
```

```python
g = sns.FacetGrid(data=tips, col="time", row="smoker")
g.map(sns.histplot, "total_bill");
```

```python
g = sns.FacetGrid(data=tips, col="time", row="smoker")
g.map(sns.histplot, "total_bill", "tip");
```

```python
g = sns.FacetGrid(data=tips, col="time", row="smoker")
g.map(plt.scatter, "total_bill", "tip");
```

```python
g = sns.FacetGrid(data=tips, col="time", row="smoker", hue="sex")
g.map(plt.scatter, "total_bill", "tip")
g.add_legend();
```
